```python
from scipy.io import arff
import pandas as pd

print(5)

# Load the .arff file into Python
data = arff.loadarff('hypothyroid.arff')
df = pd.DataFrame(data[0])

# Display the first 10 rows
df.head(10)
```

```python
# Get a value out of the dataframe
print(df["sex"])
print()
print(df["sex"][0])
```

```python
# Remove all rows which have NaN (missing value)
print("Nrows before", df.shape[0])
df_removed = df.dropna()
print("Nrows after", df_removed.shape[0])
```

```python
import numpy as np

# Replace all NaN with -10000 (Scikit-learn does not like Infinity)
df_replaced = df.fillna(-10000)

# Not necessary. But if you like you can replace all ? with "missing"
# df_replaced = df_replaced.replace(b'?', b'missing')

# Display the first 3 rows
df_replaced.head(3)
```

```python
# Scikit requires all attributes to be numeric (no strings). 
# Converts all attributes (but not the class) into dummy vairables (binary)
df_dummies = pd.get_dummies(df_replaced, columns = ['Class', 'sex', 'on thyroxine', 'query on thyroxine', 'on antithyroid medication', 'sick', 'pregnant', 'thyroid surgery', 'I131 treatment', 'query hypothyroid', 'query hyperthyroid', 'lithium', 'goitre', 'tumor', 'hypopituitary', 'psych', 'TSH measured', 'T3 measured', 'TT4 measured', 'T4U measured', 'FTI measured', 'TBG measured', 'referral source'])
df_dummies.head(5)
```

```python
from sklearn.model_selection import train_test_split

# Split into test (33%) and training set (66%)
# X is the matrix of attributes (so we drop the class) and Y is the class vector
# However, because there are 4 labels in the class, Y is a matrix with 4 dummy variables 
class_columns = ["Class_b'compensated_hypothyroid'", "Class_b'negative'", "Class_b'primary_hypothyroid'", "Class_b'secondary_hypothyroid'"]
attr_columns = list(df_dummies[[i for i in list(df_dummies.columns) if i not in class_columns]].columns)

X_train, X_test, Y_train, Y_test = train_test_split(df_dummies.drop(class_columns, axis = 1), df_dummies[class_columns], test_size=0.33, random_state=0)
X_train.shape, X_test.shape, Y_train.shape, Y_test.shape
```

```python
from sklearn import tree
from sklearn.model_selection import cross_val_score

# Run the cleaned data through a decision tree 1 time. See https://scikit-learn.org/stable/modules/tree.html
clf = tree.DecisionTreeClassifier()

# 10 fold cross-validation. See https://scikit-learn.org/stable/modules/model_evaluation.html
accuracies = cross_val_score(clf, X_train, Y_train, scoring='accuracy',  cv=10) 
print("Accuracies from 10-fold cross-validation:", accuracies)
print("Mean accuracy:", np.mean(accuracies))
```

```python
# Replace all missing values with the mean/mode
# Alternatively, see the SimpleImputer class at https://scikit-learn.org/stable/modules/impute.html
def meanOrMode(x):
    if x.dtype == "object":
        return x.mode()
    return x.mean() # Or the median?


df_imputemean = df.apply(lambda x: x.fillna(meanOrMode(x)), axis=0)
df_imputemean.head(10)
```

```python
from sklearn.preprocessing import KBinsDiscretizer

# Binning. See:
#    - https://scikit-learn.org/stable/auto_examples/preprocessing/plot_discretization_strategies.html
#    - https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.KBinsDiscretizer.html


# Apply binning to FTI
est = KBinsDiscretizer(n_bins=5, encode='ordinal', strategy='quantile')

# Split FTI into a 2D array where each element is a list with one element
FTI_split = [[x] for x in df_imputemean["FTI"]]


# Binning with equal-frequency bins
est.fit(FTI_split)  
FTI_transformed = est.transform(FTI_split)
df_imputemean_FTIbin = df_imputemean.copy()
df_imputemean_FTIbin["FTI"] = FTI_transformed
df_imputemean_FTIbin.head(10)
```

```python

# Compute the interquartile range
Q1 = X_train.quantile(0.25)
Q3 = X_train.quantile(0.75)
IQR = Q3 - Q1

# Remove outliers using the interquartile range method
df_outliersrm = X_train[~((X_train < (Q1 - 1.5 * IQR)) | (X_train > (Q3 + 1.5 * IQR))).any(axis=1)]
print("Nrows before removing outliers", X_train.shape[0])
print("Nrows after removing outliers", df_outliersrm.shape[0])
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import mutual_info_classif
from sklearn.preprocessing import LabelEncoder

# See tutorials at https://scikit-learn.org/stable/modules/feature_selection.html

# Convert class from string into integers
enc = LabelEncoder()
class_num = enc.fit_transform(df_imputemean["Class"])

# Select the k=2 best numeric features, by measuring their mutual information with the class. 
# This code only applies to numeric attributes
numeric_attributes = ["TT4", "T4U", "FTI"]
X_allNumericAttr = df_imputemean[numeric_attributes]
selector = SelectKBest(mutual_info_classif, k=2)
selector.fit(X_allNumericAttr, class_num)

# The columns to keep
kBestAttributes = selector.get_support(indices=True)
print("The best numeric attributes are", [ numeric_attributes[i] for i in kBestAttributes ])
```

```python
# This next line is important if you want to see the plot!
%matplotlib inline 
import matplotlib.pyplot as plt
from sklearn.feature_selection import RFECV
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import LabelEncoder

# Wrapping method aka recursive feature elimination (RFE)
# See tutorial https://scikit-learn.org/stable/auto_examples/feature_selection/plot_rfe_with_cross_validation.html#sphx-glr-auto-examples-feature-selection-plot-rfe-with-cross-validation-py

# Convert class from string into integers or RFECV will have a hissy fit
enc = LabelEncoder()
class_num = enc.fit_transform(df_imputemean["Class"])

# 3-fold cross validation using a decision tree. This will recursively build a tree, remove the worst features, and repeat
# until there are at least min_features_to_select=2 features left
numeric_attributes = ["TT4", "T4U", "FTI"]
X_allNumericAttr = df_imputemean[numeric_attributes]
clf = tree.DecisionTreeClassifier()
rfecv = RFECV(estimator=clf, step=1, cv=StratifiedKFold(3), scoring='accuracy')
rfecv.fit(X_allNumericAttr, class_num)

print("Optimal number of features : %d" % rfecv.n_features_)
plt.figure()
plt.xlabel("Number of features selected")
plt.ylabel("Cross validation score (nb of correct classifications)")
plt.plot(range(1, len(rfecv.grid_scores_) + 1), rfecv.grid_scores_)
plt.show()

print("The rankings of the numeric attributes are:")
for i in range(0, len(numeric_attributes)):
    print("\t", numeric_attributes[i], "is rank", rfecv.ranking_[i])
```
