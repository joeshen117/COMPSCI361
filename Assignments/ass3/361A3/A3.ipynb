# Split data into train/test 70/30-split

```python
from sklearn.model_selection import train_test_split

# Split into validation (30%) and training set (70%)
# X is the matrix of attributes (so we drop the class) and Y is the class vector
class_columns = ["target"]
attr_columns = list(df.columns)

X_train, X_test, Y_train, Y_test = train_test_split(df.drop(class_columns, axis = 1), df[class_columns], test_size=0.30, random_state=0)
X_train.shape, X_test.shape, Y_train.shape, Y_test.shape
```

# Preprocess and train model using Pipeline

```python
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.naive_bayes import BernoulliNB, GaussianNB
from sklearn.feature_selection import SelectFromModel, RFECV, RFE, SelectKBest, mutual_info_regression, mutual_info_classif
from sklearn.model_selection import StratifiedKFold
from sklearn.ensemble import ExtraTreesClassifier, RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import KBinsDiscretizer
from sklearn.preprocessing import Normalizer, StandardScaler, RobustScaler
from sklearn.svm import SVR, LinearSVC
import statistics


impute = ('imputation', SimpleImputer(strategy = "median")) # median was used to impute the missing values
normalize = ('normalize', Normalizer(norm='l2')) # no indication of outliers, prefer a more stable solution
# Using Random Forest to do feature selection, non-recursive approach, just returns all attributes above the mean importance
feature_select1 = ('feature_select1', SelectFromModel(RandomForestClassifier(n_estimators=66, random_state=None), threshold="mean"))
feature_select = ('feature_select', SelectKBest(mutual_info_classif, k = 14)) # refine selection
binning = ('binning', KBinsDiscretizer(n_bins=6, encode='ordinal', strategy='uniform')) # 6 equal width bins
nb = ("naive_bayes", BernoulliNB())

scaler = ('robustscaler',RobustScaler())

#feature_select2 = ('feature_select2', RFECV(estimator=LinearSVC(penalty="l2", dual = True, random_state = None), step=1, cv=5))
feature_select2 = ('feature_select2', RFECV(estimator=RandomForestClassifier(n_estimators=66, random_state=None), step=1, cv=5))

#feature_select = ('feature_select', SelectFromModel(LinearSVC(penalty="l2")))

#model1 = ("naive_bayes", GaussianNB())
acc_scores = []
for i in range (20):
    steps1 = [impute, normalize, feature_select1, feature_select, binning, nb]

    pipeline1 = Pipeline(steps1)
    pipeline1.fit(X_train, Y_train.values.ravel())
    prediction1 = pipeline1.predict(X_test)
    acc = accuracy_score(Y_test, prediction1)
    acc_scores.append(acc)

    print(acc)
print("average accuracy: {}".format(statistics.mean(acc_scores)))
#print("Done")
```

# GridSearchCV

```python
from sklearn.model_selection import GridSearchCV

parameters = {}
# parameters["feature_select__estimator__min_samples_split"] = [i for i in range (2,103,10)]
parameters["feature_select1__estimator__n_estimators"] = [i for i in range (50,301,20)]
#parameters["feature_select1__estimator__max_features"] = [20, 40, 60, 70, "sqrt", "log2"]
parameters["feature_select__k"] = [i for i in range (10,40,2)]
parameters["feature_select1__threshold"] = ["mean", "median", "1.25*mean", "1.25*median", "0.8*mean", "0.8*median"]
#parameters["binning__n_bins"] = [i for i in range (5,9)]
#parameters["binning__strategy"] = ["kmeans", "uniform"]

np.random.seed(seed=0)
CV = GridSearchCV(estimator = pipeline1, param_grid = parameters, cv=5, scoring = "accuracy", n_jobs = -1)
CV.fit(X_train, Y_train.values.ravel())

print(CV.best_score_)    
print(CV.best_params_)
```

```python
Y_pred = CV.predict(X_test)
accuracy_score(Y_test, Y_pred)
```

```python
from sklearn.dummy import DummyClassifier

model2 = ("dummy", DummyClassifier(strategy = "most_frequent"))
steps2 = [model2]
pipeline2 = Pipeline(steps2)
pipeline2.fit(X_train, Y_train.values.ravel())
prediction_train = pipeline2.predict(X_train)
prediction_test = pipeline2.predict(X_test)

print("Dummy Classifier")
print("Train accuracy:", accuracy_score(Y_train, prediction_train))
print("Test accuracy:", accuracy_score(Y_test, prediction_test))

```
